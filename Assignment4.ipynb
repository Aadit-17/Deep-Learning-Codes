```python
import numpy as np

np.random.seed(0)
X = np.random.randint(0, 10, size=(395, 3))
y = np.random.randint(0, 2, size=(395, 1))

W1 = np.array([[0.179, -0.186, -0.008, -0.048],
               [0.044, -0.028, -0.063, -0.131],
               [0.01, -0.035, -0.004, 0.088]])

b1 = np.zeros((1, 4))  # Bias for hidden layer
W2 = np.array([[0.088], [0.171], [0.005], [-0.04]])
b2 = np.zeros((1, 1))  # Bias for output layer

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def forward_pass(X, W1, b1, W2, b2):
    Z1 = np.dot(X, W1) + b1  # Input to hidden layer
    A1 = sigmoid(Z1)         # Activation of hidden layer
    Z2 = np.dot(A1, W2) + b2  # Input to output layer
    y_hat = sigmoid(Z2)      # Output prediction
    return Z1, A1, Z2, y_hat

def compute_loss(y, y_hat):
    m = y.shape[0]
    loss = -1/m * (np.dot(y.T, np.log(y_hat)) + np.dot((1-y).T, np.log(1-y_hat)))
    return loss

# Backpropagation
def backpropagation(X, y, Z1, A1, Z2, y_hat, W1, W2, b1, b2, learning_rate=0.01):
    m = X.shape[0]

    dZ2 = y_hat - y
    dW2 = 1/m * np.dot(A1.T, dZ2)
    db2 = 1/m * np.sum(dZ2, axis=0, keepdims=True)
    
    dA1 = np.dot(dZ2, W2.T)
    dZ1 = dA1 * A1 * (1 - A1)
    dW1 = 1/m * np.dot(X.T, dZ1)
    db1 = 1/m * np.sum(dZ1, axis=0, keepdims=True)
    
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    return W1, b1, W2, b2

def train(X, y, W1, b1, W2, b2, epochs=1000, learning_rate=0.01):
    for i in range(epochs):
        # Forward pass
        Z1, A1, Z2, y_hat = forward_pass(X, W1, b1, W2, b2)
        loss = compute_loss(y, y_hat)
        
        # Backpropagation
        W1, b1, W2, b2 = backpropagation(X, y, Z1, A1, Z2, y_hat, W1, W2, b1, b2, learning_rate)
        
        if i % 100 == 0:
            print(f"Epoch {i}, Loss: {loss[0,0]:.4f}")
    return W1, b1, W2, b2

W1, b1, W2, b2 = train(X, y, W1, b1, W2, b2, epochs=1000, learning_rate=0.01)

_, _, _, y_hat = forward_pass(X, W1, b1, W2, b2)
predictions = (y_hat > 0.5).astype(int)
accuracy = np.mean(predictions == y)
print(f"Training Accuracy: {accuracy * 100:.2f}%")
```

```python
def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(x, 0)

def softmax(x):
    expX = np.exp(x)
    return expX/np.sum(expX, axis = 0)

def derivative_tanh(x):
    return (1 - np.power(np.tanh(x), 2))

def derivative_relu(x):
    return np.array(x > 0, dtype = np.float32)
```

```python
def initialize_parameters(n_x, n_h, n_y):
    w1 = np.random.randn(n_h, n_x)*0.01
    b1 = np.zeros((n_h, 1))
    w2 = np.random.randn(n_y, n_h)*0.01
    b2 = np.zeros((n_y, 1))
    
    parameters = {
        "w1" : w1,
        "b1" : b1,
        "w2" : w2,
        "b2" : b2
    }
    return parameters

def forward_propagation(x, parameters):
    
    w1 = parameters['w1']
    b1 = parameters['b1']
    w2 = parameters['w2']
    b2 = parameters['b2']
    
    z1 = np.dot(w1, x) + b1
    a1 = tanh(z1)
    z2 = np.dot(w2, a1) + b2
    a2 = softmax(z2)
    
    forward_cache = {
        "z1" : z1,
        "a1" : a1,
        "z2" : z2,
        "a2" : a2
    }
    return forward_cache

def cost_function(a2, y):
    m = y.shape[1]
    cost = -(1/m)*np.sum(y*np.log(a2))
    return cost

def backward_prop(x, y, parameters, forward_cache):
    
    w1 = parameters['w1']
    b1 = parameters['b1']
    w2 = parameters['w2']
    b2 = parameters['b2']
    
    a1 = forward_cache['a1']
    a2 = forward_cache['a2']
    
    m = x.shape[1]
    
    dz2 = (a2 - y)
    dw2 = (1/m)*np.dot(dz2, a1.T)
    db2 = (1/m)*np.sum(dz2, axis = 1, keepdims = True)
    dz1 = (1/m)*np.dot(w2.T, dz2)*derivative_tanh(a1)
    dw1 = (1/m)*np.dot(dz1, x.T)
    db1 = (1/m)*np.sum(dz1, axis = 1, keepdims = True)
    
    gradients = {
        "dw1" : dw1,
        "db1" : db1,
        "dw2" : dw2,
        "db2" : db2
    }
    return gradients

def update_parameters(parameters, gradients, learning_rate):
    
    w1 = parameters['w1']
    b1 = parameters['b1']
    w2 = parameters['w2']
    b2 = parameters['b2']
    
    dw1 = gradients['dw1']
    db1 = gradients['db1']
    dw2 = gradients['dw2']
    db2 = gradients['db2']
    
    w1 = w1 - learning_rate*dw1
    b1 = b1 - learning_rate*db1
    w2 = w2 - learning_rate*dw2
    b2 = b2 - learning_rate*db2
    
    parameters = {
        "w1" : w1,
        "b1" : b1,
        "w2" : w2,
        "b2" : b2
    }
    return parameters
```

```python
def model(x, y, n_h, learning_rate, iterations):
    
    n_x = x.shape[0]
    n_y = y.shape[0]
    
    cost_list = []
    
    parameters = initialize_parameters(n_x, n_h, n_y)
    
    for i in range(iterations):
        forward_cache = forward_propagation(x, parameters)
        cost = cost_function(forward_cache['a2'], y)
        gradients = backward_prop(x, y, parameters, forward_cache)
        parameters = update_parameters(parameters, gradients, learning_rate)
       
        cost_list.append(cost)
        if(i%(iterations/10) == 0 or i == iterations-1):
            print("Cost after", i, "iterations is :", cost)
    return parameters, cost_list

iterations = 100
n_h = 1000
learning_rate = 0.02
Parameters, Cost_list = model(X_train, Y_train, n_h = n_h, learning_rate = learning_rate, iterations = iterations)
```

#### Implement the following collection of GD based optimization algorithms used for deep learning

#### Consider a model of linear function y=f(x)=w0+ w1*x

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([5, 15, 25, 35, 45, 55])
y = np.array([5, 20, 14, 32, 22, 38])

epochs = 100
alpha = 0.001 

def compute_mse(x, y, w0, w1):
    predictions = w0 + w1 * x
    return np.mean((predictions - y) ** 2)

# Gradient Clipping Function
def clip_gradients(grad, threshold=1.0):
    if np.abs(grad) > threshold:
        grad = np.sign(grad) * threshold
    return grad

# Stochastic Gradient Descent
def sgd(x, y, alpha, epochs):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            w0 -= alpha * grad_w0
            w1 -= alpha * grad_w1
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# Stochastic Gradient Descent + Momentum
def sgd_momentum(x, y, alpha, epochs, momentum=0.9):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    v0, v1 = 0, 0
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            v0 = momentum * v0 + alpha * grad_w0
            v1 = momentum * v1 + alpha * grad_w1
            w0 -= v0
            w1 -= v1
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# Adagrad
def adagrad(x, y, alpha, epochs, epsilon=1e-8):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    G0, G1 = 0, 0
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            G0 += grad_w0 ** 2
            G1 += grad_w1 ** 2
            w0 -= (alpha / (np.sqrt(G0) + epsilon)) * grad_w0
            w1 -= (alpha / (np.sqrt(G1) + epsilon)) * grad_w1
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# RMSprop
def rmsprop(x, y, alpha, epochs, beta=0.9, epsilon=1e-8):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    E0, E1 = 0, 0
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            E0 = beta * E0 + (1 - beta) * grad_w0 ** 2
            E1 = beta * E1 + (1 - beta) * grad_w1 ** 2
            w0 -= (alpha / (np.sqrt(E0) + epsilon)) * grad_w0
            w1 -= (alpha / (np.sqrt(E1) + epsilon)) * grad_w1
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# AdaDelta
def adadelta(x, y, alpha, epochs, beta=0.9, epsilon=1e-8):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    E0, E1 = 0, 0
    delta_w0, delta_w1 = 0, 0
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            
            # Update running averages of the squared gradients
            E0 = beta * E0 + (1 - beta) * grad_w0 ** 2
            E1 = beta * E1 + (1 - beta) * grad_w1 ** 2
            
            # Avoid sqrt of negative values
            delta_w0_update = (np.sqrt(delta_w0 + epsilon) / np.sqrt(E0 + epsilon)) * grad_w0
            delta_w1_update = (np.sqrt(delta_w1 + epsilon) / np.sqrt(E1 + epsilon)) * grad_w1
            
            # Update parameters
            w0 -= delta_w0_update
            w1 -= delta_w1_update
            
            # Update running averages of the parameter updates
            delta_w0 = beta * delta_w0 + (1 - beta) * delta_w0_update ** 2
            delta_w1 = beta * delta_w1 + (1 - beta) * delta_w1_update ** 2
            
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# Adam
def adam(x, y, alpha, epochs, beta1=0.9, beta2=0.999, epsilon=1e-8):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    m0, m1 = 0, 0
    v0, v1 = 0, 0
    errors = []
    for epoch in range(epochs):
        for i in range(len(x)):
            prediction = w0 + w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            m0 = beta1 * m0 + (1 - beta1) * grad_w0
            m1 = beta1 * m1 + (1 - beta1) * grad_w1
            v0 = beta2 * v0 + (1 - beta2) * grad_w0 ** 2
            v1 = beta2 * v1 + (1 - beta2) * grad_w1 ** 2
            m0_hat = m0 / (1 - beta1 ** (epoch + 1))
            m1_hat = m1 / (1 - beta1 ** (epoch + 1))
            v0_hat = v0 / (1 - beta2 ** (epoch + 1))
            v1_hat = v1 / (1 - beta2 ** (epoch + 1))
            w0 -= alpha * m0_hat / (np.sqrt(v0_hat) + epsilon)
            w1 -= alpha * m1_hat / (np.sqrt(v1_hat) + epsilon)
        errors.append(compute_mse(x, y, w0, w1))
    return errors

# Nesterov Accelerated Gradient (NAG)
def nag(x, y, alpha, epochs, momentum=0.9):
    w0, w1 = np.random.randn() * 0.01, np.random.randn() * 0.01
    v0, v1 = 0, 0
    errors = []
    for epoch in range(epochs):
        # Calculate the look-ahead position
        lookahead_w0 = w0 - momentum * v0
        lookahead_w1 = w1 - momentum * v1
        
        for i in range(len(x)):
            # Use look-ahead weights for prediction
            prediction = lookahead_w0 + lookahead_w1 * x[i]
            grad_w0 = clip_gradients((prediction - y[i]))
            grad_w1 = clip_gradients((prediction - y[i]) * x[i])
            
            # Update velocities
            v0 = momentum * v0 + alpha * grad_w0
            v1 = momentum * v1 + alpha * grad_w1
            
            # Update weights
            w0 -= v0
            w1 -= v1
            
        errors.append(compute_mse(x, y, w0, w1))
    return errors

sgd_errors = sgd(x, y, alpha, epochs)
sgd_momentum_errors = sgd_momentum(x, y, alpha, epochs)
adagrad_errors = adagrad(x, y, alpha, epochs)
rmsprop_errors = rmsprop(x, y, alpha, epochs)
adadelta_errors = adadelta(x, y, alpha, epochs)
adam_errors = adam(x, y, alpha, epochs)
nag_errors = nag(x, y, alpha, epochs)

plt.figure(figsize=(10, 6))
plt.plot(sgd_errors, label='SGD')
plt.plot(sgd_momentum_errors, label='SGD with Momentum')
plt.plot(adagrad_errors, label='Adagrad')
plt.plot(rmsprop_errors, label='RMSprop')
plt.plot(adadelta_errors, label='Adadelta')
plt.plot(adam_errors, label='Adam')
plt.plot(nag_errors, label='Nesterov Accelerated Gradient (NAG)')
plt.xlabel('Epochs')
plt.ylabel('Mean Squared Error')
plt.title('Comparison of Optimization Algorithms with Clipped Gradients')
plt.legend()
plt.show()
```

#### Q1 Construct a ML Model using Neural network and Display the Training / Testing Loss

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

np.random.seed(42)
X = np.random.rand(100, 10)
y = 3 * X[:, 0] + 2 * X[:, 1] + np.random.randn(100) * 0.1 

# ReLU Activation Function and its derivative
def relu(Z):
    return np.maximum(0, Z)

def relu_derivative(Z):
    return Z > 0

# Linear Activation Function (Identity)
def linear(Z):
    return Z

def linear_derivative(Z):
    return np.ones_like(Z)

# Initialize parameters for the neural network
def initialize_parameters(input_size, hidden_size, output_size):
    W1 = np.random.randn(hidden_size, input_size) * 0.01
    b1 = np.zeros((hidden_size, 1))
    W2 = np.random.randn(output_size, hidden_size) * 0.01
    b2 = np.zeros((output_size, 1))
    
    return W1, b1, W2, b2

# Forward propagation
def forward_propagation(X, W1, b1, W2, b2):
    Z1 = np.dot(W1, X) + b1
    A1 = relu(Z1)
    Z2 = np.dot(W2, A1) + b2
    A2 = linear(Z2)
    
    return Z1, A1, Z2, A2

# Backpropagation
def back_propagation(X, Y, Z1, A1, Z2, A2, W2):
    m = X.shape[1]
    
    dZ2 = A2 - Y
    dW2 = (1/m) * np.dot(dZ2, A1.T)
    db2 = (1/m) * np.sum(dZ2, axis=1, keepdims=True)
    dA1 = np.dot(W2.T, dZ2)
    dZ1 = dA1 * relu_derivative(Z1)
    dW1 = (1/m) * np.dot(dZ1, X.T)
    db1 = (1/m) * np.sum(dZ1, axis=1, keepdims=True)
    
    return dW1, db1, dW2, db2

# Update parameters using gradient descent
def update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate):
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2
    
    return W1, b1, W2, b2

# Training the neural network
def train(X, Y, hidden_size, learning_rate, epochs):
    input_size = X.shape[0]
    output_size = 1
    W1, b1, W2, b2 = initialize_parameters(input_size, hidden_size, output_size)
    
    for i in range(epochs):
        Z1, A1, Z2, A2 = forward_propagation(X, W1, b1, W2, b2)
        cost = np.mean((A2 - Y) ** 2)  # Mean Squared Error
        dW1, db1, dW2, db2 = back_propagation(X, Y, Z1, A1, Z2, A2, W2)
        W1, b1, W2, b2 = update_parameters(W1, b1, W2, b2, dW1, db1, dW2, db2, learning_rate)
        
        if i % 100 == 0:
            print(f"Cost after iteration {i}: {cost}")
    
    return W1, b1, W2, b2

def predict(X, W1, b1, W2, b2):
    _, _, _, A2 = forward_propagation(X, W1, b1, W2, b2)
    return A2

# Standardize the dataset
scaler = StandardScaler()
X = scaler.fit_transform(X).T
y = y.reshape(1, -1)

# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X.T, y.T, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test = X_train.T, X_test.T, y_train.T, y_test.T

hidden_size = 10
learning_rate = 0.01
epochs = 1000

W1, b1, W2, b2 = train(X_train, y_train, hidden_size, learning_rate, epochs)

y_pred_train = predict(X_train, W1, b1, W2, b2)
y_pred_test = predict(X_test, W1, b1, W2, b2)

# Calculate Mean Squared Error
train_mse = np.mean((y_pred_train - y_train) ** 2)
test_mse = np.mean((y_pred_test - y_test) ** 2)

print(f"Train MSE: {train_mse}")
print(f"Test MSE: {test_mse}")
```

#### Q2 Perform K-Fold Validation for the IRIS Dataset and visualize the overfit and Underfit

```python
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
import numpy as np
from sklearn.model_selection import validation_curve
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold

np.random.seed(0)
plt.style.use('ggplot')

iris = load_iris()
X, y = iris.data, iris.target

kf = KFold(n_splits=20)
list_training_error = []
list_testing_error = []

for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    model = MLPRegressor()
    model.fit(X_train, y_train)
    
    y_train_data_pred = model.predict(X_train)
    y_test_data_pred = model.predict(X_test)
    
    fold_training_error = mean_absolute_error(y_train, y_train_data_pred)
    fold_testing_error = mean_absolute_error(y_test, y_test_data_pred)
    
    list_training_error.append(fold_training_error)
    list_testing_error.append(fold_testing_error)

plt.subplot(1,2,1)
plt.plot(range(1, kf.get_n_splits() + 1), np.array(list_training_error).ravel(), 'o-')
plt.xlabel('Number of Folds')
plt.ylabel('Training Error')
plt.title('Training Error across Folds')

plt.subplot(1,2,2)
plt.plot(range(1, kf.get_n_splits() + 1), np.array(list_testing_error).ravel(), 'o-')
plt.xlabel('Number of Folds')
plt.ylabel('Testing Error')
plt.title('Testing Error across Folds')
plt.show()
```
